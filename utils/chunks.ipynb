```python
import pandas as pd

data = pd.read_csv('data/combined_data.csv')
```

```python
# Let's take only first and last chunks
# 200 tokens from the beginig of the text and 300 tokens from the end
new_texts = data.text.apply(lambda text: ' '.join(text.split(' ')[:200]+text.split(' ')[-300:]))
data['text'] =  new_texts
data.to_csv('data/first_last_chunks.csv')
```

```python
# Let's take only the conclusion of the text
new_texts = data.text.apply(lambda text: ' '.join(text.split(' ')[-500:]))
data['text'] =  new_texts
data.to_csv('data/last_chunk.csv')
```
